```python
from flask import Flask, jsonify, request

app = Flask(__name__)

# Smartphone class demonstrating Object-Oriented style
class Smartphone:
    def __init__(self, brand, storage, memory):
        self.brand = brand
        self.storage = storage
        self.memory = memory

    # Method to convert object attributes to a dictionary for JSON responses
    def to_dict(self):
        return {
            "brand": self.brand,
            "storage": self.storage,
            "memory": self.memory
        }

# Pre-defined list of smartphone objects
phones = [
    Smartphone("iphone", "128GB", "6GB"),
    Smartphone("samsung", "256GB", "8GB"),
    Smartphone("huawei", "512GB", "12GB")
]

# Endpoint 1: GET All (Displays all 3 phones)
@app.route('/phones', methods=['GET'])
def get_all_phones():
    return jsonify([phone.to_dict() for phone in phones])

# Endpoint 2: GET with parameter (Filter by brand, e.g., /phone?brand=samsung)
@app.route('/phone', methods=['GET'])
def get_phone_by_brand():
    brand_query = request.args.get('brand')
    
    if not brand_query:
        return jsonify({"error": "Please provide a brand parameter, e.g., ?brand=samsung"}), 400

    # Search for matching brand (case-insensitive)
    phone_found = next((p.to_dict() for p in phones if p.brand.lower() == brand_query.lower()), None)

    if phone_found:
        return jsonify(phone_found)
    else:
        return jsonify({"message": "Smartphone not found"}), 404

if __name__ == '__main__':
    # use_reloader=False prevents the ipykernel conflict
    app.run(debug=True, host='127.0.0.1', port=5200, use_reloader=False)
```
